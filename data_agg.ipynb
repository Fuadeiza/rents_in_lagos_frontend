```python
import pandas as pd
import streamlit as st



df = pd.read_json("output.json")

df.head()
```

```python
df
df.columns
```

```python
df = df.groupby(["house_type","latitude","longitude"]).mean()
df = df.reset_index()
# df.columns = df.columns.str.strip()

df = df[df.house_type != 0]
df = df[df.house_type <7]

df.tail(20)
```

```python
df.size
```

```python
import leafmap.foliumap as leafmap



m = leafmap.Map(center=[6.5244, 3.3792], zoom=10)
m.add_heatmap(
    df,
    latitude="latitude",
    longitude="longitude",
    value="price",
    name="Heat map",
    radius=20,
)


m
```
